```python
import torch
from embedding import PositionalEncoding, Embedding
```

```python
L = 20
sigma = 0.1
embed_size = 32
```

```python
part_idx = L // 2
spins = torch.zeros(L // 2)
```

```python
pos_enc = PositionalEncoding(embed_size=embed_size, max_chain_len=L)
```

```python
# seq, batch, embed
empty_chains = torch.zeros(L, 1, embed_size)
pos1, pos2 = pos_enc(empty_chains, empty_chains)
```

```python
assert torch.allclose(pos1, pos2)
```

```python
import matplotlib.pyplot as plt

# Reshape the pos1 and pos2 tensors to remove the batch dimension
pos1_reshaped = pos1.squeeze(1)
pos2_reshaped = pos2.squeeze(1)

# Plot the heatmap for pos1
plt.figure(figsize=(8, 6))
plt.imshow(pos1_reshaped, cmap="hot", interpolation="nearest")
plt.title("Attention is All You Need - Positional Encoding (pos1)")
plt.xlabel("Embedding Dimension")
plt.ylabel("Sequence Length")
plt.colorbar()
plt.show()

# Plot the heatmap for pos2
plt.figure(figsize=(8, 6))
plt.imshow(pos2_reshaped, cmap="hot", interpolation="nearest")
plt.title("Attention is All You Need - Positional Encoding (pos2)")
plt.xlabel("Embedding Dimension")
plt.ylabel("Sequence Length")
plt.colorbar()
plt.show()
```
